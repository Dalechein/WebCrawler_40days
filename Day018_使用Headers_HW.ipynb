## 作業目標

利用開發者工具去尋找「 https://www.zhihu.com/explore 」和「 http://aicoin.cn 」網站的 HTTP 網址，以及他所帶的 Headers 分別是哪些，試著用 request 去發送有沒有帶 Headers 的差異。


### 1. https://www.zhihu.com/explore 

```python
import requests
url = 'https://www.zhihu.com/explore'
r = requests.get(url)
r.encoding = 'utf-8'
print(r.text)
```

### 利用查看開發者工具觀察 Headers


```python
import requests
url = 'http://www.zhihu.com'

headers = {
    'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64; rv:86.0) Gecko/20100101 Firefox/86.0'
}

r = requests.get(url, headers=headers)

r.encoding = 'utf-8'
print(r.text[0:600])
```

### 2. http://aicoin.cn

```python
import requests

url = 'http://aicoin.cn/'
r = requests.get(url)
r.encoding = 'utf-8'
print(r.text)
```

### 利用查看開發者工具觀察 Headers


```python
# 帶header
import requests
url = 'http://aicoin.cn/'

headers = {
    'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64; rv:86.0) Gecko/20100101 Firefox/86.0'
}

r = requests.get(url, headers=headers)

print(r.text[0:600])
```

```python
#其他案例，模擬手機
import requests
url = 'http://aicoin.cn/'

headers = { 
    'User-Agent':'Mozilla/5.0 (Linux; U; Android 4.3; en-us; SM-N900T Build/JSS15J) AppleWebKit/534.30 (KHTML, like Gecko) Version/4.0 Mobile Safari/534.30'
}

r = requests.get(url, headers=headers)

print(r.text[0:600])
```

### 3. https://movies.yahoo.com.tw/api/v1/areas_by_movie_theater

```python
# 指定ID，沒有帶 Headers 
movie_id = 11072

url = 'https://movies.yahoo.com.tw/api/v1/areas_by_movie_theater'
payload = {'movie_id':str(movie_id)}

resp = requests.get(url, params=payload)
#print(resp.json())  # 若有需要，列印出json原始碼
#print(resp)  # 若有需要，列印出json原始碼
# 這裡回傳的格式是JSON格式的資料，要解析JSON擷取資料
for p in resp.json():
    print('放映地區: {0}, 代號(area_id): {1}'.format(p['title'], p['area_id']))
```

```python
# 指定ID，帶 Headers，只有 User-Agent 
movie_id = 11072

url = 'https://movies.yahoo.com.tw/api/v1/areas_by_movie_theater'
payload = {'movie_id':str(movie_id)}

headers = {
    'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64; rv:86.0) Gecko/20100101 Firefox/86.0'
}

resp = requests.get(url, params=payload, headers=headers)
#print(resp.json())  # 若有需要，列印出json原始碼
#print(resp)  # 若有需要，列印出json原始碼
# 這裡回傳的格式是JSON格式的資料，要解析JSON擷取資料
for p in resp.json():
    print('放映地區: {0}, 代號(area_id): {1}'.format(p['title'], p['area_id']))
```
