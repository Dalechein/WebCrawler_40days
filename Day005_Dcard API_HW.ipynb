```python
from selenium import webdriver
from selenium.webdriver.support.ui import Select
from bs4 import BeautifulSoup
import time
import json
```

```python
# 1. é€™å€‹ API ä¸€æ¬¡æœƒå›å‚³å¹¾ç­†è³‡æ–™ï¼Ÿæ¯ä¸€ç­†è³‡æ–™åŒ…å«å“ªäº›æ¬„ä½ï¼Ÿ
browser = webdriver.Chrome(executable_path='chromedriver')
browser.get("https://www.dcard.tw/_api/forums/pet/posts?popular=true")

time.sleep(7)

# å–å¾—è³‡æ–™ï¼ˆç­‰ç€è¦½å™¨è³‡æ–™å‡ºç¾å¾Œæ‰åŸ·è¡Œï¼‰
html_source = browser.page_source
html = BeautifulSoup(html_source, 'lxml')
data = json.loads(html.text)
```

```python
print('ä¸€æ¬¡å‚³ï¼š%s ç­†' % len(data))
data[:3]
```

```python
# 2. å–å‡ºæ¯ä¸€ç­†è³‡æ–™çš„ã€Œæ¨™é¡Œã€ã€ã€Œè²¼æ–‡æ™‚é–“ã€ã€ã€Œç•™è¨€äººæ•¸ã€ã€ã€ŒæŒ‰è®šäººæ•¸ã€
posts = dict()

for d in data:
    posts[d['id']]= [d['title']]+[d['createdAt']]+[str(d['commentCount'])]+[str(d['likeCount'])]
    print('id:', d['id'])
    print('Info:',posts[d['id']])
#print(posts.items())
```

```python
# 3. è¨ˆç®—ç†±é–€/éç†±é–€æ–‡ç« çš„ã€Œå¹³å‡ç•™è¨€äººæ•¸ã€èˆ‡ã€Œå¹³å‡æŒ‰è®šäººæ•¸ã€
#è¨ˆç®—ç†±é–€æ–‡ç« çš„ã€Œå¹³å‡ç•™è¨€äººæ•¸ã€èˆ‡ã€Œå¹³å‡æŒ‰è®šäººæ•¸ã€
commentTotal = 0
likeTotal = 0
for v, w in posts.items():
    commentTotal += int(w[2])
    likeTotal += int(w[3])
print('å¹³å‡ç•™è¨€äººæ•¸: ', commentTotal/len(data))
print('å¹³å‡æŒ‰è®šäººæ•¸: ', likeTotal/len(data))
```
