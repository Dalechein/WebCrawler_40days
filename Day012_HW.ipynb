### ① 取出今天所有的發文

```python
import requests
from bs4 import BeautifulSoup

url = 'https://www.ettoday.net/news/news-list.htm'
r = requests.get(url)

soup = BeautifulSoup(r.text, "html5lib")

for d in soup.find(class_="part_list_2").find_all('h3'):
    print(d.find_all('a')[0].text)
```

### ② 如果想要依照類別分類，怎麼儲存會比較好？

```python
url = 'https://www.ettoday.net/news/news-list.htm'
r = requests.get(url)

soup = BeautifulSoup(r.text, "html5lib")
ClassNews = dict()

for d in soup.find(class_="part_list_2").find_all('h3'):   
    if d.find('em').text not in ClassNews.keys():
        ClassNews[d.find('em').text] = [d.find_all('a')[0].text.replace(u'\u3000',u' ')]
    else:
        ClassNews[d.find('em').text] = ClassNews[d.find('em').text] + [d.find_all('a')[0].text.replace(u'\u3000',u' ')]
```
